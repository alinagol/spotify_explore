```python
import requests
import json
from requests_oauthlib import OAuth1
import base64
import six
from pprint import pprint
import pandas as pd
```

# Authorize

```python
config = json.load(open('config/access.json'))

auth_header = base64.b64encode(str(config['Client ID'] + ':' + config['Client Secret']).encode('ascii'))

headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'Authorization': 'Basic '+ auth_header.decode('ascii')
}

payload = { 'grant_type': 'client_credentials'}

auth_req = requests.post('https://accounts.spotify.com/api/token', headers=headers, data=payload)
```

```python
headers = {
    'Authorization': 'Bearer ' + auth_req.json()['access_token']
}

s = requests.Session()
s.headers.update(headers)
```

# Get an Artist's Top Tracks

```python
artist_id = '1Xyo4u8uXC1ZmMpatF05PJ'

headers['Content-Type'] = 'application/json'

payload = {'country': 'NL'}

tracks = s.get('https://api.spotify.com/v1/artists/%s/top-tracks'%(artist_id), headers=headers, params=payload).json()['tracks']
```

```python
tracksDF = pd.DataFrame()
```

```python
for track in tracks:
    
    tr = s.get('https://api.spotify.com/v1/audio-features/%s'%(track['id']), headers=headers).json()
    tr.update({'name': track['name'], 'popularity': track['popularity']})
    tracksDF = tracksDF.append(tr, ignore_index=True)
```
